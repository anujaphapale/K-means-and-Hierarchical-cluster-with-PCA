```python
import pandas as pd 
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale 
```

```python
wine= pd.read_csv("wine.csv")
wine.head()
```

```python
wine['Type'].value_counts()
```

```python
# Considering only numerical data 
wine.data = wine.iloc[:,1:]
wine.data.head()
```

```python
# Converting into numpy array
wine_ary = wine.data.values
wine_ary
```

```python
# Normalizing the numerical data 
wine_normal = scale(wine_ary)
```

```python
wine_normal
```

```python
# Applying PCA Fit Transform to dataset
pca = PCA()
pca_values = pca.fit_transform(wine_normal)
pca_values
```

```python
# PCA Components matrix or convariance Matrix
pca.components_
```

```python
# The amount of variance that each PCA explains is
var = pca.explained_variance_ratio_
var
```

```python
# Cumulative variance 
var1 = np.cumsum(np.round(var,decimals = 4)*100)
var1
```

```python
pca.components_
```

```python
# Variance plot for PCA components obtained 
plt.plot(var1,color="red")
```

```python
pca_values[:,0:1]
```

```python
# plot between PCA1 and PCA2 
x = pca_values[:,0:1]
y = pca_values[:,1:2]
#z = pca_values[:2:3]
plt.scatter(x,y)
```

```python
finalDf = pd.concat([wine['Type'],pd.DataFrame(pca_values[:,0:3],columns=['PC1','PC2','PC3'])], axis = 1)
```

# k mean clustering

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans
```

```python
# How to find optimum number of  cluster
#The K-means algorithm aims to choose centroids that minimise the inertia, or within-cluster sum-of-squares criterion:
```

```python
finalDf.head()
```

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i,random_state=0)
    kmeans.fit(finalDf)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
#Build Cluster algorithm
from sklearn.cluster import KMeans
clusters_new = KMeans(3, random_state=42)
clusters_new.fit(finalDf)
```

```python
clusters_new.labels_
```

```python
k_FinalDf=finalDf.copy()
k_FinalDf.head()
```

```python
#Assign clusters to the data set
k_FinalDf['clusterid_new'] = clusters_new.labels_
k_FinalDf.head()
```
